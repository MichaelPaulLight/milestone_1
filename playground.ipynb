## generate column names

```python

def generate_column_mappings(dataframes):
    """
    Generates descriptive and consistent column names for each dataframe in the given dictionary.
    
    Parameters:
    dataframes (dict): A dictionary where the keys are table names and values are pandas dataframes.
    
    Returns:
    dict: A dictionary with table names as keys and lists of new descriptive column names as values.
    """
    column_name_mapping = {}
    
    for table_name, df in dataframes.items():
        original_columns = df.columns
        new_columns = []
        
        for col in original_columns:
            # Clean up the column name
            new_col = col.lower()  # Convert to lowercase
            new_col = new_col.replace('%', 'percentage')  # Replace percentage symbol
            # new_col = new_col.replace(r'perc', 'percentage')
            new_col = new_col.replace(' ', '_')  # Replace spaces with underscores
            new_col = re.sub(r'\W+', '_', new_col)  # Replace non-word characters with underscores
            
            # Standardize 'county' columns
            if 'county' in new_col:
                new_col = 'county_name'
            if 'unnamed__0' in new_col:
                new_col = 'id'
            # Remove trailing underscores
            new_col = new_col.strip('_')
            
            # Add the cleaned up column name to the list
            new_columns.append(new_col)
        
        # Add the mapping for this table
        column_name_mapping[table_name] = new_columns
    
    return column_name_mapping
```
